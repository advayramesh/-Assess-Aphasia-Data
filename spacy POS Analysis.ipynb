```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")
DOCNAME = nlp("Apple is looking at buying U.K. startup for 1 billion")

# parts of speech tagging
# for token in DOCNAME:
#     print(((token.text,
#             token.lemma_,
#             token.pos_,
#             token.tag_,
#             token.dep_,
#             token.shape_,
#             token.is_alpha,
#             token.is_stop,
#             )))
# for token in doc:
#     print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
#             token.shape_, token.is_alpha, token.is_stop)

```

```python
#Initialize array (columns for excel)

Group = []
Type = []
Key = []
Text = []
Freq = []

Debug = 1;
```

```python
#Group Read
group = "LHD#"
```

```python
# To print the text, Part of Speech, Dependancy of tokens
# for token in DOCNAME:
#     print(token.text,token.pos_, token.dep_)

#Part of speech where k = key value assigned by spaCy and f = frequency and DOCNAME is whatever you have called your document
POS_counts = DOCNAME.count_by(spacy.attrs.POS)
POS_counts


if ( Debug == 1 ):
  print("Part of speech and frequency")
  for k, f in sorted(POS_counts.items()):
    print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')

for k, f in sorted(POS_counts.items()):
  #print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')
  Group.append(group)
  Type.append('POS')
  Key.append(k)
  Text.append(DOCNAME.vocab[k].text)
  Freq.append(f)


#Dependencies where k = key value and f = frequency and DOCNAME is whatever you have called your document
DEP_counts = DOCNAME.count_by(spacy.attrs.DEP)
DEP_counts

if ( Debug == 1 ):
  print("Dependencies and frequency")
  for k, f in sorted(DEP_counts.items()):
    print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')


for k, f in sorted(DEP_counts.items()):
  #print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')
  Group.append(group)
  Type.append("DEP")
  Key.append(k)
  Text.append(DOCNAME.vocab[k].text)
  Freq.append(f)
```

```python
# Write to excel file
import pandas as pd
Dataset = pd.DataFrame()
Dataset['Group'] = Group
Dataset['Type'] = Type
Dataset['Key'] = Key
Dataset['Text'] = Text
Dataset['Freq'] = Freq


Dataset
#Dataset.to_csv('Aphasia_POS_DEP_Freq.csv')
```
