```python
#Initialize array (columns for excel)

Group = []
Type = []
Key = []
Text = []
Freq = []

LHD = ""
RHD = ""

NGroup = []
Nountext = []
NFreq = []

Debug = 0;
```

```python
def pos_dep_parsing(chunk):
# To print the text, Part of Speech, Dependancy of tokens
# for token in DOCNAME:
#     print(token.text,token.pos_, token.dep_)

  DOCNAME = nlp(chunk)
  #Part of speech where k = key value assigned by spaCy and f = frequency and DOCNAME is whatever you have called your document
  POS_counts = DOCNAME.count_by(spacy.attrs.POS)
  POS_counts


  if ( Debug == 1 ):
    print("Part of speech and frequency")
    for k, f in sorted(POS_counts.items()):
      print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')

  for k, f in sorted(POS_counts.items()):
    #print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')
    Group.append(group)
    Type.append('POS')
    Key.append(k)
    Text.append(DOCNAME.vocab[k].text)
    Freq.append(f)


  #Dependencies where k = key value and f = frequency and DOCNAME is whatever you have called your document
  DEP_counts = DOCNAME.count_by(spacy.attrs.DEP)
  DEP_counts

  if ( Debug == 1 ):
    print("Dependencies and frequency")
    for k, f in sorted(DEP_counts.items()):
      print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')


  for k, f in sorted(DEP_counts.items()):
    #print(f'{k}. {DOCNAME.vocab[k].text:{10}}:{f}')
    Group.append(group)
    Type.append("DEP")
    Key.append(k)
    Text.append(DOCNAME.vocab[k].text)
    Freq.append(f)
```

```python
def get_noun_frequency(text):
    # Process the text with spaCy
    doc = nlp(text)

    # Create a Counter to store noun frequencies
    noun_counter = Counter()

    # Iterate over tokens in the processed document
    for token in doc:
        # (pos == 'NN' or pos=='NNP' or pos=='NNS' or pos=='NNPS'
        if token.pos_ == "NOUN" or token.pos_ == "PROPN":
            noun_counter[token.text.lower()] += 1

    for noun, frequency in noun_counter.items():
      #print(f"Noun: {noun}, Frequency: {frequency}")
      NGroup.append(group)
      Nountext.append(noun)
      NFreq.append(frequency)

    #return noun_counter

# Example text
#text = "The quick brown fox jumped over the lazy dog. The dog barked loudly. fox fox foxy"

# Get noun frequency using the function
#noun_frequency = get_noun_frequency(text)

# Print the noun frequency
#for noun, frequency in noun_frequency.items():
#    print(f"Noun: {noun}, Frequency: {frequency}")
```

```python
#Read file
def read_and_split_file(file_path, delimiter):
    with open(file_path, 'r') as file:
        content = file.read()
        chunks = content.split(delimiter)
        return chunks

file_path = "/content/Cookie Theft all speech sub v2.txt"
delimiter = "----------------------------------------------------"
type_len = 6
lhd = 0
rhd = 0
others = 0

chunks = read_and_split_file(file_path, delimiter)
for index, chunk in enumerate(chunks):
  chunk = chunk.strip()
  type = chunk[:type_len]
  #print({type})
  describe = (chunk[type_len:].strip())
  if ( "LHD" in type ):
    lhd = lhd + 1
    LHD = LHD + describe
    group = type #"LHD"+ str(index+1)
  elif ("RHD" in type ):
    rhd = rhd + 1
    RHD = RHD + describe
    group = type #"RHD"+ str(index+1)
  else:
    others = others + 1
  pos_dep_parsing(describe)
  get_noun_frequency(describe)

```

```python
#Do all LHD, RHD as full group

group = "LHDAll"
pos_dep_parsing(LHD)
get_noun_frequency(LHD)
group = "RHDAll"
pos_dep_parsing(RHD)
get_noun_frequency(RHD)
```
