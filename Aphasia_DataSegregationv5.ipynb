```python
Text = []
Class = []
Noun_count = []
RNoun_count = []
Verb_count = []
RVerb_count = []
Determiner_count = []
RDeterminer_count = []
Adjective_count = []
RAdjective_count = []
Uhh_count = []
RUhh_count = []
Symbol_count = []
NumWord_count = []
Pause = []
NumPause = []
AphasiaType = []
Humor = []
Sigh = []
Pointing = []
Searching = []
Object = []
ProperNoun = []
Pauses = []
RPauses = []
Stutter = []
RStutter = []
RPauseStutter = []
Motor = []
Remember = []
Sentences = []
NoMeaning = []
PurposeObj = []
RPurposeObj = []
PurposeNN = []
RPurposeNN = []
PurposeVerb = []
RPurposeVerb = []
RPurposeWord = []
Prompt = []

for i in range(0,len(data['annotations'])):
  pausepos = data['annotations'][i][0].find("(pause T")
  fsentences = data['annotations'][i][0][0:pausepos]
  sentences = nltk.sent_tokenize(fsentences)
  #sentences = nltk.sent_tokenize(data['annotations'][i][0])
  Text.append(sentences)

  class_ = data['annotations'][i][1]['entities'][0][2]
  pause = data['annotations'][i][1]['entities'][1][0]
  numpause = data['annotations'][i][1]['entities'][2][0]
  aphasiatype = data['annotations'][i][1]['entities'][5][2]
  humor = data['annotations'][i][1]['entities'][6][0]
  sigh = data['annotations'][i][1]['entities'][7][0]
  pointing = data['annotations'][i][1]['entities'][8][0]
  searching = data['annotations'][i][1]['entities'][9][0]
  objects = data['annotations'][i][1]['entities'][10][0]
  propernn = data['annotations'][i][1]['entities'][11][0]
  pauses = data['annotations'][i][1]['entities'][12][0]
  stutter = data['annotations'][i][1]['entities'][13][0]
  motor = data['annotations'][i][1]['entities'][14][0]
  remember = data['annotations'][i][1]['entities'][15][0]
  sent = data['annotations'][i][1]['entities'][16][0]
  nomean = data['annotations'][i][1]['entities'][17][0]
  purposeobj = data['annotations'][i][1]['entities'][18][0]
  purposenn = data['annotations'][i][1]['entities'][19][0]
  purposevrb = data['annotations'][i][1]['entities'][20][0]
  prompt = data['annotations'][i][1]['entities'][21][0]

  noun = []
  verb = []
  dt = []
  adjective = []
  numword = []
  uhh = []
  sym = []
  for sentence in sentences:
    for word,pos in nltk.pos_tag(nltk.word_tokenize(str(sentence))):
      if ( "uh" in word.lower() or "hmm" in word.lower() or "mhm" in word.lower()  ):
        uhh.append(word + " - " + "UH")
      elif ("th" == word.lower()):
        dt.append(word + " - " + "DT")
      elif(pos == 'NN' or pos=='NNP' or pos=='NNS' or pos=='NNPS'):
        noun.append(word)
      elif (pos == 'VB' or pos == 'VBD' or pos == 'VBG' or pos == 'VBN' or pos == 'VBP'):
        verb.append(word)
      elif (pos == 'DT'):
        dt.append(word)
      elif (pos == 'JJ' or pos == 'JJR' or pos == 'JJS'):
        adjective.append(word)
      elif (pos == 'UH'):
        uhh.append(word)
      elif (pos == 'SYM'):
        sym.append(word)
      numword.append(word)



  wc = len(numword)
  Noun_count.append(len(noun))
  RNoun_count.append(len(noun)*100/wc)
  Verb_count.append(len(verb))
  RVerb_count.append(len(verb)*100/wc)
  Determiner_count.append(len(dt))
  RDeterminer_count.append(len(dt)*100/wc)
  Adjective_count.append(len(adjective))
  RAdjective_count.append(len(adjective)*100/wc)
  Uhh_count.append(len(uhh))
  RUhh_count.append(len(uhh)*100/wc)
  NumWord_count.append(len(numword))
  Symbol_count.append(len(sym))
  Class.append(class_)
  Pause.append(pause)
  NumPause.append(numpause)
  AphasiaType.append(aphasiatype)
  Humor.append(humor)
  Sigh.append(sigh)
  Pointing.append(pointing)
  Searching.append(searching)
  Object.append(object)
  ProperNoun.append(propernn)
  Pauses.append(pauses)
  RPauses.append(pauses*100/wc)
  Stutter.append(stutter)
  RStutter.append(stutter*100/wc)
  RPauseStutter.append(pause+stutter)
  Motor.append(motor)
  Remember.append(remember)
  Sentences.append(sent)
  NoMeaning.append(nomean)
  PurposeObj.append(purposeobj)
  RPurposeObj.append(purposeobj*100/10)
  PurposeNN.append(purposenn)
  RPurposeNN.append(purposenn*100/6)
  PurposeVerb.append(purposevrb)
  RPurposeVerb.append(purposevrb*100/8)
  RPurposeWord.append((purposeobj + purposenn+ purposevrb)/24)
  Prompt.append(prompt)


```

```python
import pandas as pd
Dataset = pd.DataFrame()
Dataset['Text'] = Text
Dataset['Noun_Count'] = Noun_count
Dataset['RNoun_Count'] = RNoun_count
Dataset['Verb_count'] = Verb_count
Dataset['RVerb_count'] = RVerb_count
Dataset['Determiner'] = Determiner_count
Dataset['RDeterminer'] = RDeterminer_count
Dataset['Adjective'] = Adjective_count
Dataset['RAdjective'] = RAdjective_count
Dataset['Uhh'] = Uhh_count
Dataset['RUhh'] = RUhh_count
Dataset['Symbol'] = Symbol_count
Dataset['Total Word'] = NumWord_count
Dataset['Total Pause'] = Pause
Dataset['Number Of Pause'] = NumPause
Dataset['Aphasia Type'] = AphasiaType
Dataset['Humor'] = Humor
Dataset['Sigh'] = Sigh
Dataset['Pointing'] = Pointing
Dataset['Searching'] = Searching
Dataset['Object'] = Object
Dataset['ProperNoun'] = ProperNoun
Dataset['Pauses'] = Pauses
Dataset['RPauses'] = RPauses
Dataset['Stutter'] = Stutter
Dataset['RStutter'] = RStutter
Dataset['PauseStutter'] = RPauseStutter
Dataset['Motor'] = Motor
Dataset['Remember'] = Remember
Dataset['Sentences'] = Sentences
Dataset['NoMeaning'] = NoMeaning
Dataset['PurposeObj'] = PurposeObj
Dataset['RPurposeObj'] = RPurposeObj
Dataset['PurposeNN'] = PurposeNN
Dataset['RPurposeNN'] = RPurposeNN
Dataset['PurposeVerb'] = PurposeVerb
Dataset['RPurposeVerb'] = RPurposeVerb
Dataset['RPurposeWord'] = RPurposeWord
Dataset['Prompt'] = Prompt

Dataset['Class'] = Class
```
